```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Dropping irrelevant columns

```python
traindf.drop(columns=[ 'LotFrontage','Alley','FireplaceQu', 'GarageType','GarageYrBlt', 'GarageFinish','GarageQual',
                      'GarageCond','EnclosedPorch', '3SsnPorch', 'ScreenPorch','PoolQC','Fence', 'MiscFeature','MiscVal','YrSold'],inplace = True)
```

```python
traindf.head()
```

```python
traindf.info()
```

```python
#dropping the null values
traindf.dropna(inplace = True)
```

```python
#check for duplicates
traindf.duplicated().sum()
```

```python
#let's create a dataframe for the 'Id' column since we gonna use it for submission
trainId = traindf['Id']
traindf = traindf.drop('Id',axis = 1)
```

```python
#let's create a list of all numerical columns
numeric_cols = traindf.select_dtypes(include=['int64']).columns
numeric_cols
```

```python
traindf.shape
```

# Removing outliers

```python
for column in numeric_cols:
    q1 = traindf[column].quantile(0.25)
    q3 = traindf[column].quantile(0.75)
    iqr = q3-q1
    upper = q3 + (1.5*iqr)
    lower = q1 - (1.5*iqr)
    traindf = traindf[(traindf[column] >= lower) & (traindf[column] <= upper)]
```

```python
traindf.shape
```

```python
sns.boxplot(x=traindf['BedroomAbvGr'])
```

# Importing scikit-learn libraries and ML algorithms

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import(
LabelEncoder,
MinMaxScaler,
PolynomialFeatures
)
from sklearn.model_selection import(
KFold,
train_test_split,
GridSearchCV,
)
from sklearn.feature_selection import f_regression,SelectKBest
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

### Converting categorical to numerical

```python
le = LabelEncoder()
for column in traindf.columns:
    if traindf[column].dtype == 'object':
        traindf[column] = le.fit_transform(traindf[column])
        
traindf
```

# Separating input features from target feature and doing feature selection

```python
X = traindf.drop('SalePrice',axis =1)
y = traindf['SalePrice']
```

### Performing feature selection

```python
fs = SelectKBest(score_func = f_regression , k = 15)
fs.fit(X,y)
fs.scores_
```

```python
features = pd.DataFrame(X.columns)
score = pd.DataFrame(fs.scores_)
feature_score = pd.concat([features,score],axis = 1)
feature_score.columns = ['feature','f_score']
print(feature_score.nlargest(15,columns = 'f_score'))
```

```python
X.columns
```

```python
X.drop(columns = ['MSSubClass', 'MSZoning', 'LotArea', 'Street', 'LotShape',
                 'LandContour', 'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood',
                 'Condition1', 'Condition2', 'BldgType', 'HouseStyle',
                 'OverallCond','RoofStyle', 'RoofMatl',
                 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'MasVnrArea',
                 'ExterCond','BsmtCond', 'BsmtExposure',
                 'BsmtFinType1', 'BsmtFinSF1', 'BsmtFinType2', 'BsmtFinSF2',
                 'BsmtUnfSF', 'Heating', 'CentralAir', 'Electrical','2ndFlrSF', 'LowQualFinSF', 'BsmtFullBath',
                 'BsmtHalfBath','HalfBath', 'BedroomAbvGr', 'KitchenAbvGr','Functional', 'Fireplaces', 'PavedDrive', 'WoodDeckSF', 'OpenPorchSF', 'PoolArea',
                 'MoSold', 'SaleType', 'SaleCondition'],inplace = True)
```

### Splitting data into sub-train and sub-test for evaluating model performance

```python
X_subtrain,X_subtest,y_subtrain,y_subtest = train_test_split(X,y,test_size=0.3)
```

```python
X_subtrain
```

### Scaling the data,we gonna use standard scaler since most of the distributions are normal

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
scaler = StandardScaler()
scaler.fit(X_subtrain)
X_subtrain = scaler.transform(X_subtrain)
X_subtest = scaler.transform(X_subtest)
```

### Building various models

```python
model1 = RandomForestRegressor(n_estimators = 10,criterion = 'poisson')
model1.fit(X_subtrain,y_subtrain)
prediction = model1.predict(X_subtest)
print(np.sqrt(mean_squared_error(y_subtest,prediction)))
```

```python
linear = LinearRegression()
linear.fit(X_subtrain,y_subtrain)
linear_prediction = linear.predict(X_subtest)
print(np.sqrt(mean_squared_error(y_subtest,linear_prediction)))
```

### Linear Regression has the best score

```python
linear.score(X_subtest,y_subtest)
```
